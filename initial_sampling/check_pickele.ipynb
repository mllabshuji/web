```python
def connect_dicts(dict1, dict2, save_path = None):
    merged_dict = {}
    for key in dict1.keys(): # å„keyã”ã¨ã«çµåˆ 
        merged_dict[key] = pd.concat([dict1[key], dict2[key]],
                                      axis=0, ignore_index=True)
        merged_dict = pd.DataFrame(merged_dict)
    if save_path: 
        with open(save_path, 'wb') as out_dir:
            pickle.dump(merged_dict, out_dir)
    return merged_dict

def connect_initial_sampling(random_list, rad_list, save = True):
    if type(random_list) != list:
        random_list = [random_list]
    if type(rad_list) != list:
        rad_list = [rad_list]
    for random in random_list:
        for rad in rad_list:
            for i in range(0, 4):
                with open(f'Random{random}_rad{rad}_to{i}.pkl', 'rb') as f1:
                    add_dict = pickle.load(f1)
                if i == 0:
                    ALL_dict = add_dict
                else:
                    ALL_dict = connect_dicts(ALL_dict, add_dict)
            if save:
                with open(f'Random{random}_rad{rad}_ALL.pkl', "wb") as f:
                    pickle.dump(ALL_dict, f)
    return ALL_dict

def check_dict(dict):
    for key, value in dict.items():
        print(f"Key: {key}, Type of value: {type(value)}, data_num: {value.shape}")
```
