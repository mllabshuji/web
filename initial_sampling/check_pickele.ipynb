```python
import numpy as np
import os
import sys
import pickle
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import re
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr, spearmanr
import shutil
```

```python
def connect_dicts(dict1, dict2, save_path = None):
    merged_dict = {}
    for key in dict1.keys(): # 各keyごとに結合 
        merged_dict[key] = pd.concat([dict1[key], dict2[key]],
                                      axis=0, ignore_index=True)
        merged_dict = pd.DataFrame(merged_dict)
    if save_path: 
        with open(save_path, 'wb') as out_dir:
            pickle.dump(merged_dict, out_dir)
    return merged_dict

def connect_initial_sampling(random_list, rad_list, save = True):
    if type(random_list) != list:
        random_list = [random_list]
    if type(rad_list) != list:
        rad_list = [rad_list]
    for random in random_list:
        for rad in rad_list:
            for i in range(0, 4):
                with open(f'Random{random}_rad{rad}_to{i}.pkl', 'rb') as f1:
                    add_dict = pickle.load(f1)
                if i == 0:
                    ALL_dict = add_dict
                else:
                    ALL_dict = connect_dicts(ALL_dict, add_dict)
            if save:
                with open(f'Random{random}_rad{rad}_ALL.pkl', "wb") as f:
                    pickle.dump(ALL_dict, f)
    return ALL_dict

def check_dict(dict):
    for key, value in dict.items():
        print(f"Key: {key}, Type of value: {type(value)}, data_num: {value.shape}")
```

```python
# compliance 比較グラフ
os.makedirs("Result_graph", exist_ok=True)
for random_num in range(2,7):
    for rad in [0.5, 0.75]:
        data_frame = pd.DataFrame(data_frame_co)
        columns = [col for col in data_frame.columns if f"Random{random_num}_rad{rad}" in col][::-1]
        g = sns.pairplot(data_frame[columns], kind="reg", plot_kws={"line_kws": {"color": "red"}})

        g.figure.suptitle(f"Random{random_num}_rad{rad}_compliance", fontsize=40)
        g.figure.subplots_adjust(top=0.95)

        min350 = data_frame[f"Random{random_num}_rad{rad}_mesh350_compliance"].min()
        max350 = data_frame[f"Random{random_num}_rad{rad}_mesh350_compliance"].max()

        # 全体のプロット（g）に対して、対角以外に同じ軸範囲で y = x を描画
        for i in range(g.axes.shape[0]):
            for j in range(g.axes.shape[1]):
                if i != j:
                    ax = g.axes[i, j]
                    # ax.set_xlim(min350, max350)
                    # ax.set_ylim(min350, max350)
                    ax.set_xlim(min350*0.7, max350*1.0)
                    ax.set_ylim(min350*0.7, max350*1.0)

                    ax.plot([min350, max350], [min350, max350], color='red', linestyle='--')
        for i, col in enumerate(columns):
            label = re.sub(f"Random{random_num}_rad{rad}_", "", col).replace("_compliance", "")
            g.axes[i, 0].set_ylabel(label, fontsize=14)
            g.axes[-1, i].set_xlabel(label, fontsize=14)

        plt.savefig(f"Result_graph/Random{random_num}_rad{rad}_compliance.pdf", bbox_inches='tight')
        plt.close()
```
